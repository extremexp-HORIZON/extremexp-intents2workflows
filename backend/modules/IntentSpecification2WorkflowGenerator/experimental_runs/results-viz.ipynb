```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
prev_bf_gen = pd.read_csv('prev-bruteforce-runs/prev-bruteforce-results/prev-bruteforce-results.csv')

prev_bf_gen.rename(columns={
    'time': 'old_bf_time',
    'num_workflows': 'old_bf_num_workflows',
}, inplace=True)
prev_bf_gen = prev_bf_gen.drop("overlap", axis='columns')

prev_bf_gen
```

```python
new_bf_gen = pd.read_csv('new-bruteforce-runs/new-bruteforce-results/new-bruteforce-results.csv')

new_bf_gen.rename(columns={
    'time': 'new_bf_time',
    'num_workflows': 'new_bf_num_workflows',
}, inplace=True)
new_bf_gen = new_bf_gen.drop("overlap", axis='columns')

new_bf_gen
```

```python
new_mid_case = pd.read_csv('mid-case-runs/mid-case-results/mid-case-results.csv')

new_mid_case.rename(columns={
    'time': 'new_mc_time',
    'num_workflows': 'new_mc_num_workflows',
}, inplace=True)
new_mid_case = new_mid_case.drop("overlap", axis='columns')

new_mid_case
```

```python
new_best_case = pd.read_csv('best-case-runs/best-case-results/best-case-results.csv')

new_best_case.rename(columns={
    'time': 'new_bc_time',
    'num_workflows': 'new_bc_num_workflows',
}, inplace=True)
new_best_case = new_best_case.drop("overlap", axis='columns')

new_best_case
```

```python
results_df = pd.merge(prev_bf_gen, new_bf_gen, on=['comps', 'reqs_per_comp', 'comps_per_req'])

results_df = pd.merge(results_df, new_mid_case, on=['comps', 'reqs_per_comp', 'comps_per_req'])

results_df = pd.merge(results_df, new_best_case, on=['comps', 'reqs_per_comp', 'comps_per_req'])

results_df
```

```python
results_df['c_rpc_cpr'] = results_df.apply(lambda row: f"({int(row['comps'])}, {int(row['reqs_per_comp'])}, {int(row['comps_per_req'])})", axis=1)

results_df
```
